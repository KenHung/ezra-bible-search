## Extract Terms

Extract terms from pattern: XXX = "XXX" and remove some invalid data

```python
eq_terms = extract_term(cbol_dict.defs)
'  '.join(eq_terms.dropna())
```

```python
cbol_dict['term'] = eq_terms.str.replace('[或．•‧,/\-，、\[\]]', ' ').str.split()
'  '.join(cbol_dict.term.dropna().explode())
```

Terms that didn't fall into the XXX = "XXX" pattern

```python
not_match = cbol_dict[cbol_dict.defs.str.contains('(?:專有|人名|地名)') & ~cbol_dict.defs.str.contains('n([^n]*?) ?=')]
defs_list = not_match.defs.str.split(';')
not_match['defs2:'] = defs_list.apply(lambda x: x[2:])
not_match
```

```python
term_2nd = not_match.defs.str.replace('\\r\\n\\r\\n', ')', regex=False)\
                         .str.replace('\\r\\n', ' ', regex=False)\
                         .str.extractall('(\d\) ?([^\d\s]+))+')[1]
term_2nd_to_clean = term_2nd.str.contains(
    '(?:磐石|毀滅|欽定本|女|兒子|旋轉|領域|陽性|的|基督徒|施洗者|詞|字意|敵人|原先為海|使徒|上帝|比喻|群集|破裂|橄欖|禮貌性稱呼|草原居民|用於擬人法|鄉村|鴕鳥|去|離去|忙於|比喻用法|樹枝|石頭|沉入水裡|靜止|力量|堅固|(冰)|精疲力盡|耶和華|園|驚恐|草原|重要大事|種族名稱|偉大|指下埃及地區|神自己|盼望|法老|巨人|一|角|市|;|[a-z])'
    ) | (term_2nd.str.len() > 6)
print('Extract: ' + '|'.join(term_2nd[~term_2nd_to_clean]))
print()
print('Clean: ' + '  '.join(term_2nd[term_2nd_to_clean]))
```

```python
term_2nd_clean = term_2nd[~term_2nd_to_clean]
term_2nd_clean = term_2nd_clean.drop((8150, 1)).drop((3573, 1)).reset_index(level=[1])

cbol_dict.term.fillna(term_2nd_clean[1], inplace=True)
'  '.join(cbol_dict.term.dropna().explode())
```

## Add POS

```python
cbol_dict['term_pos'] = cbol_dict.term.where(cbol_dict.term.isna(), 'nz')
cbol_dict.term_pos.where(~cbol_dict.defs.str.contains('人名'), 'nr', inplace=True)
cbol_dict.term_pos.where(~cbol_dict.defs.str.contains('(?:父親|兒子|後裔|一位|一名|家族|的人)'), 'nr', inplace=True)
cbol_dict.term_pos.where(~cbol_dict.defs.str.contains('地名'), 'ns', inplace=True)
cbol_dict[cbol_dict.term.notna()].term_pos.value_counts()
```

## Generate Terms File

```python
from hanziconv import HanziConv

bible_terms = cbol_dict[cbol_dict.term.notna()][['term', 'term_pos']].explode('term')

bible_terms['term'] = bible_terms.term.str.strip('婦族人的,"')\
                                      .apply(HanziConv.toSimplified)
bible_terms.drop_duplicates('term').to_csv('bible_terms.txt', index=False, header=None, sep=' ')
```
